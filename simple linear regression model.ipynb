```python
#importing required libraries 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#impoting data set
salary_data = pd.read_csv("Salary_Data.csv")
```

```python
#visualizing first 5 rows of our data set
salary_data.head()
```

```python
#plotting yearsExperience column (barplot)
salary_data["YearsExperience"].plot(figsize=(10,5))
```

```python
#plotting Salary column (barplot)
salary_data["Salary"].plot(figsize=(10,5))
```

```python
#another way of plotting data
fig, ax = plt.subplots(2,1,figsize=(10,5))
salary_data["YearsExperience"].plot(kind="bar",figsize=(10,5),ax=ax[0])
salary_data["Salary"].plot(kind="bar",figsize=(10,5),ax=ax[1])
```

```python
#since both are linear we can use linear reggressor
from sklearn.linear_model import LinearRegression
```

```python
#creating a linear regression model
model = LinearRegression()
```

```python
#importing train_test_split to split data into training set and testing set
from sklearn.model_selection import train_test_split
```

```python
#creating input variable
X = salary_data.drop('Salary', axis=1)
```

```python
#creating output variable
y = salary_data["Salary"]
```

```python
#splitting our data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
# training our model with training set
model.fit(X_train, y_train)
```

```python
#checking the performance of our model using training set itself
model.score(X_train, y_train)
```

```python
#predicting output using trained model, here salary is predicted using years of experience
preds = model.predict(X_test)
```

```python
#importing evaluation metrics to evaluate our model
from sklearn.metrics import r2_score
```

```python
#evaluating using r2_score
r2_score(y_test,preds)
```
